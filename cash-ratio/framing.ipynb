## Time Resampling

```python
df = pd.read_csv('../dummy/bri-data-hackathon-cr-optimization/train.csv', index_col='periode', parse_dates=True)
```

```python
# get mean from all columns over the year
# df.resample(rule='A').mean()
```

### Kas Kantor

```python
# Resample Monthly Basis
df['kas_kantor'].resample(rule='M').mean().plot.bar(color=list('rgbkymc'))
```

```python
df['cash_in_kantor'].resample(rule='M').mean().plot.bar(color=list('rgbkymc'))
```

```python
df['cash_out_kantor'].resample(rule='M').mean().plot.bar(color=list('rgbkymc'))
```

### Kas Echannel

```python
df['kas_echannel'].resample(rule='M').mean().plot.bar(color=list('rgbkymc'))
```

```python
df['cash_in_echannel'].resample(rule='M').mean().plot.bar(color=list('rgbkymc'))
```

```python
df['cash_out_echannel'].resample(rule='M').mean().plot.bar(color=list('rgbkymc'))
```

## Moving Averages

### Kas Kantor

```python
df['kas_kantor'].plot(figsize=(12,6))
# compare with monthly mean average (the orange line)
df['kas_kantor'].rolling(window=30).mean().plot(figsize=(12,6))
# df['kas_kantor'].expanding().mean().plot(figsize=(12,6))
```

```python
df['cash_in_kantor'].plot(figsize=(12,6))
# compare with monthly mean average (the orange line)
df['cash_in_kantor'].rolling(window=30).mean().plot(figsize=(12,6))
```

```python
df['cash_out_kantor'].plot(figsize=(12,6))
# compare with monthly mean average (the orange line)
df['cash_out_kantor'].rolling(window=30).mean().plot(figsize=(12,6))
```
